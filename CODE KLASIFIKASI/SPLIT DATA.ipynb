```python
import os

def count_files_in_subfolders(folder_path):
    # Mengecek apakah folder_path ada
    if not os.path.exists(folder_path):
        print("Folder tidak ditemukan!")
        return
    
    # Loop untuk setiap subfolder di dalam folder_path
    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)
        
        # Pastikan yang dibaca adalah folder
        if os.path.isdir(subfolder_path):
            # Menghitung jumlah file dalam subfolder
            file_count = len([f for f in os.listdir(subfolder_path) 
                              if os.path.isfile(os.path.join(subfolder_path, f))])
            print(f"Folder '{subfolder}' memiliki {file_count} file.")

# Ganti path ini dengan path ke folder asl_alphabet_train Anda
folder_path = "asl_alphabet_train"
count_files_in_subfolders(folder_path)
```

```python
import os
import random
import shutil

def split_dataset(folder_path, train_ratio=0.8, val_ratio=0.1, test_ratio=0.1):
    # Mengecek apakah folder_path ada
    if not os.path.exists(folder_path):
        print("Folder tidak ditemukan!")
        return
    
    # Menentukan folder tujuan
    train_folder = os.path.join(folder_path, 'train')
    val_folder = os.path.join(folder_path, 'val')
    test_folder = os.path.join(folder_path, 'test')
    
    # Membuat folder tujuan jika belum ada
    for folder in [train_folder, val_folder, test_folder]:
        if not os.path.exists(folder):
            os.makedirs(folder)
    
    # Loop untuk setiap subfolder di dalam folder_path
    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)
        
        # Pastikan yang dibaca adalah folder
        if os.path.isdir(subfolder_path):
            # Dapatkan semua file dalam subfolder
            files = [f for f in os.listdir(subfolder_path) 
                     if os.path.isfile(os.path.join(subfolder_path, f))]
            
            # Hitung jumlah data untuk training, validation, dan testing
            total_files = len(files)
            train_count = int(total_files * train_ratio)
            val_count = int(total_files * val_ratio)
            test_count = total_files - train_count - val_count
            
            # Acak urutan file
            random.shuffle(files)
            
            # Pisahkan file untuk train, validation, dan test
            train_files = files[:train_count]
            val_files = files[train_count:train_count + val_count]
            test_files = files[train_count + val_count:]
            
            # Membuat subfolder untuk masing-masing kelas di folder tujuan
            for folder in [train_folder, val_folder, test_folder]:
                class_folder = os.path.join(folder, subfolder)
                if not os.path.exists(class_folder):
                    os.makedirs(class_folder)
            
            # Pindahkan file ke folder masing-masing
            for file in train_files:
                shutil.move(os.path.join(subfolder_path, file), os.path.join(train_folder, subfolder, file))
            for file in val_files:
                shutil.move(os.path.join(subfolder_path, file), os.path.join(val_folder, subfolder, file))
            for file in test_files:
                shutil.move(os.path.join(subfolder_path, file), os.path.join(test_folder, subfolder, file))
            
            print(f"Folder '{subfolder}' telah dibagi menjadi {train_count} untuk training, {val_count} untuk validation, dan {test_count} untuk testing.")

# Ganti path ini dengan path ke folder asl_alphabet_train Anda
folder_path = "asl_alphabet_train"
split_dataset(folder_path)
```
